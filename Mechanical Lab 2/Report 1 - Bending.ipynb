```python
import numpy as np
from openpyxl import load_workbook
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Young's Modulus

```python
wb = load_workbook("/Users/kunwoosmac/Desktop/Fall 2023/MMAE 419/Bending/Bending Test_BB_F2022(1).xlsx")

name = 'Steel'
fig_path = '/Users/kunwoosmac/Desktop/Fall 2023/MMAE 419/Bending/Figures'
ws = wb[name]

delta = [ws.cell(i,8).value for i in range(11, 28)]
delta = np.array(delta)

pl33i = [ws.cell(i,9).value for i in range(11, 28)]

# pl33i : [N/mm] -> pl33i/1000 : [kN/mm]
pl33i = np.array(pl33i)/1000 # Changed pl33i into kN/mm


plt.figure(figsize = (10,5))
plt.title("Steel, $P$ - $\delta$ relation (method 1)", fontsize = 17)
plt.scatter(delta, pl33i, label = 'experimental data')
plt.ylabel("PL$^3$/3I [kN / mm]", fontsize = 15)
plt.xlabel("$\delta$ [mm]", fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

reg = LinearRegression(fit_intercept =False).fit(delta.reshape(-1,1),pl33i.reshape(-1,1))

reg_a1 = reg.coef_[0][0]

plt.plot(delta.reshape(-1,1), reg.predict(delta.reshape(-1,1)), c = 'r', label = 'curve fitted data')
print("From Method 1 : E ={} kN/mm^2 = {} GPa".format(reg_a1, reg_a1*(1000**3)/10**9))
plt.grid()
plt.legend()
plt.savefig(fig_path+"/"+name+", method1.png")
plt.show()

# ========= # ========= # ========= # ========= # ========= # ========= # =========

sigma1 = [ws.cell(i,10).value for i in range(11, 28)]
sigma2 = [ws.cell(i,11).value for i in range(11, 28)]

sigma = sigma1 + sigma2

sigma1 = np.array(sigma1).reshape(-1,1); sigma2 = np.array(sigma2).reshape(-1,1)

sigma = np.array(sigma) # [N/mm^2]

epsilon1 = [ws.cell(i,3).value for i in range(11, 28)]
epsilon2 = [ws.cell(i,5).value for i in range(11, 28)]

epsilon = epsilon1 + epsilon2

epsilon1 = np.array(epsilon1).reshape(-1,1); epsilon2 = np.array(epsilon2).reshape(-1,1)


plt.figure(figsize = (10,5))
plt.title("Steel, $\sigma$ - $\epsilon$ relation (method 2)", fontsize = 17)
plt.scatter(epsilon, sigma, label = 'experimental data')
plt.ylabel("$\sigma$ [$N/mm^2$]", fontsize = 15)
plt.xlabel("$\epsilon$", fontsize = 15)

reg1 = LinearRegression(fit_intercept =False).fit(epsilon1, sigma1)
reg2 = LinearRegression(fit_intercept =False).fit(epsilon2, sigma2)

reg1_a1 = reg1.coef_[0][0]
reg2_a1 = reg2.coef_[0][0]

plt.plot(epsilon1, reg1.predict(epsilon1), c = 'orange', label = 'curve fitted data from top (compression)')
plt.plot(epsilon2, reg2.predict(epsilon2), c = 'r', label = 'curve fitted data from bottom (tension)')

plt.axhline(0, c = 'black')
plt.axvline(0, c = 'black')
plt.grid()
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

plt.legend()
plt.savefig(fig_path+"/"+name+", method2.png")
print("From Method 2 : E ={} N/mm^2 = {} Gpa".format((reg1_a1 + reg2_a1)/2, (reg1_a1 + reg2_a1)/2*(1000**2)/10**9))
print("Top : {} GPa, Bottom : {} GPa".format(reg1_a1*(1000**2)/10**9, reg2_a1*(1000**2)/10**9))
```

# Poisson's Ratio

```python
name = 'Steel'
r = 28
fig_path = '/Users/kunwoosmac/Desktop/Fall 2023/MMAE 419/Bending/Figures'
ws = wb[name]

elt = np.array([ws.cell(i,3).value for i in range(11, r)]).reshape(-1,1)
ett = -np.array([ws.cell(i,4).value for i in range(11, r)]).reshape(-1,1)
elb = np.array([ws.cell(i,5).value for i in range(11, r)]).reshape(-1,1)
etb = -np.array([ws.cell(i,6).value for i in range(11, r)]).reshape(-1,1)


plt.figure(figsize = (10,5))
plt.title(name+", -$\epsilon_t$ vs $\epsilon_l$", fontsize = 17)
plt.scatter(elt, ett, label = 'experimental data - Top')
plt.scatter(elb, etb, label = 'experimental data - Bottom', c = 'green')

plt.ylabel("$-\epsilon_t$", fontsize = 15)
plt.xlabel("$\epsilon_l$", fontsize = 15)

reg1 = LinearRegression(fit_intercept =False).fit(elt, ett)
reg2 = LinearRegression(fit_intercept =False).fit(elb, etb)

reg1_a1 = reg1.coef_[0][0]
reg2_a1 = reg2.coef_[0][0]

plt.plot(elt, reg1.predict(elt), c = 'orange', label = 'curve fitted data from top (compression)')
plt.plot(elb, reg2.predict(elb), c = 'r', label = 'curve fitted data from bottom (tension)')

plt.axhline(0, c = 'black')
plt.axvline(0, c = 'black')
plt.grid()
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

plt.legend()
plt.savefig(fig_path+"/"+name+", poisson.png")
print("Poisson : {}".format((reg1_a1 + reg2_a1)/2))
print("Top : {}, Bottom : {}".format(reg1_a1, reg2_a1))
```
