```python
# Strain gauge readings vs. measured pressure in a single graph
```

```python
Pm = np.array([ws.cell(i,2).value for i in range(8, 25)])
e45 = np.array([ws.cell(i,3).value for i in range(8, 25)])
el = np.array([ws.cell(i,4).value for i in range(8, 25)])
e_45 = np.array([ws.cell(i,5).value for i in range(8, 25)])
```

```python
plt.figure(figsize = (9, 5))
plt.title("Measured Pressure vs. Strain Gauge Readings", fontsize = 18)

plt.plot(e45, Pm, label = '$P_m vs. \epsilon_{+45}$')
plt.plot(el, Pm, label = '$P_m vs. \epsilon_l$')
plt.plot(e_45, Pm, label = '$P_m vs. \epsilon_{-45}$')

plt.ylabel("Meaured Pressure ($P_m$) [psi]", fontsize = 15)
plt.xlabel("Strain Gauge Readings", fontsize = 15)

plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)

plt.legend(fontsize = 14, loc = 'lower right')

plt.savefig(figpath + "strain vs presure")
```
