```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = np.array([20, 300, 341, 658, 1100, 2000, 2300,8750, 71000])
w_thirds = w **(1/3)

p = np.array([1000, 185, 378, 300, 190, 312, 240, 193, 65])

k = list(p*w_thirds)

k = sum(k)/len(k)
print("Average k:", k)
```

```python
# try guessing the p with w
p_guess = k / w_thirds

# Get the error
err = 100 * (abs(p_guess - p) / p)
for i, e in enumerate(err):
    print("Error {} : {}%".format(i+1, e.round(2)))
```

```python
import matplotlib.pyplot as plt

t = np.array([7, 14, 21, 28, 35, 42])
P = np.array([8, 41, 133, 250, 280, 297])

c1 = list(P/t)
print(sum(c1)/len(c1))

plt.scatter(t,P)
plt.xlabel("t (days)")
plt.ylabel("P (number of observed flies)")
```

```python
pred = 5.73*t
err = (pred - P)/P * 100
print(sum(abs(err))/len(err))
print((pred - P)/P * 100)
```

```python
plt.scatter(t,P, label = 'Real')
plt.plot(t, pred, 'r', label = 'Prediction')
plt.xlabel("t (days)")
plt.ylabel("P (number of observed flies)")
plt.legend()
```

```python
lnP = np.log(P)
lnP
```

```python
plt.scatter(t,lnP, label = 'Real')
# plt.plot(t, pred, 'r', label = 'Prediction')
plt.xlabel("t (days)")
plt.ylabel("ln P (ln (number of observed flies) ")
plt.legend()
```

```python
from sklearn.linear_model import LinearRegression

t = np.array(t)
lnP = np.array(lnP)
t = t.reshape(len(t), 1); lnP = lnP.reshape(len(lnP), 1)

reg = LinearRegression().fit(t, lnP)

reg_a0 = reg.intercept_[0]
reg_a1 = reg.coef_[0][0]

plt.scatter(t,lnP, label = 'Real')
plt.plot(t, reg.predict(t), c = 'r', label = "Fit")
plt.legend()

print("Fitted model: ln(P) = {}t + {}".format(reg_a1, reg_a0))
```
