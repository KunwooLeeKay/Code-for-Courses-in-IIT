```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
P = np.array([ [0.25, 0.25, 0.5], [0.1, 0.3, 0.6], [0.05, 0.15, 0.8] ])
P_T = P.transpose()

for i in range(10):
    P_temp = P_T
    P_T = P_T @ P_T
    if (P_temp == P_T).all():
        break

print(P_T)
```

```python
a0 = 2; b0 = 13
an = [a0]; bn = [b0]
for i in range(100):
    an.append(an[i]*0.35 + 0.1 * bn[i])
    bn.append(an[i]*0.65 + 0.9 * bn[i])
    
n = list(range(101))
plt.plot(n, an, label = "Lake A")
plt.plot(n, bn, label = "Lake B")
plt.legend()
```

```python
P = np.array([ [0,1/3, 1/3, 1/3, 0], [0,0,0,0,1], [0,1/2,0,0,1/2], [0,0,1/2, 0, 1/2], [1,0,0,0,0] ])
P = P.transpose()
eigval, eigvec = np.linalg.eig(P)
eigvec = np.hsplit(eigvec, 5)

for i in range(5):
    print("eigen value: {}\neigenvector : {}".format(eigval[i], eigvec[i].ravel()))
    print("-"*60)
```
