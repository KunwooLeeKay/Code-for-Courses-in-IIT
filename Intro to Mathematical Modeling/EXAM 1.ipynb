```python
x = np.array([10, 15, 22, 77, 142, 283, 448])
y = np.array([22.5, 36.5, 68.7, 433, 1075, 3066, 6015])

# lets write x* as x_, y* as y_, c* as c_
x_ = np.log(x)
y_ = np.log(y)

# Set what to minimize. from [r, alpha, c_] we are minimizing r. So [1,0,0]
c = np.array([1,0,0])

# Define A
A = np.zeros((14,3))
for i, xi in zip(range(0, 14, 2), x_):
    A[i] = np.array([-1, -xi, -1])
    A[i+1] = np.array([-1, xi, 1])

# Define b
b = np.zeros((14,1))
for i, yi in zip(range(0, 14, 2), y_):
    b[i] = np.array(-yi)
    b[i+1] = np.array(yi)

r_bounds = (None, None)
a_bounds = (None, None)
c_bounds = (None, None)

res = linprog(c, A_ub = A, b_ub = b, bounds=[r_bounds, a_bounds, c_bounds])

print("Minimum r : {}, alpha : {}, c* : {}".format(res['x'][0],res['x'][1],res['x'][2]))
```
