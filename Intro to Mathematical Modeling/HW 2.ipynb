# Plot the different coefficient

```python
k1 = -0.3; k2 = 0.2; k3 = 0.002; k4 = -0.001

On = [40]; Mn = [30]

n = list(range(1,1000))

for i in range(len(n) - 1):
    
    if (1+k1)*On[i] + k3*On[i]*Mn[i] < 1 or (1+k2)*Mn[i] + k4*On[i]*Mn[i] < 1:
        n = list(range(1,i+2))
        break
        
    On.append((1+k1)*On[i] + k3*On[i]*Mn[i])
    Mn.append((1+k2)*Mn[i] + k4*On[i]*Mn[i])

plt.figure(figsize = (10,5))
plt.title("k1: {}, k2: {}, k3: {}, k4: {}".format(k1,k2,k3,k4))
plt.plot(n, On, c = 'b', label = 'Owl')
plt.plot(n, Mn, c = 'g', label = 'Mice')
plt.legend()

dominator = 'Owl'
if On[-1] < Mn[-1]:
    dominator = 'Mice'

print("{} dominate after {} days".format(dominator, n[-1]))
```

```python
# Define possible coefficients
k1 = [0.18, 0.2, 0.22]
k2 = [0.27, 0.3, 0.33]
k3 = [0.00099, 0.001, 0.0011]
k4 = [0.0018, 0.002, 0.0022]

# Create possible set of combination : k1, k2, k3, k4
combination = []
for a in k1:
    for b in k2:
        for c in k3:
            for d in k4:
                combination.append(tuple([a,b,c,d]))

row = 0 
col = 0

# Plot the data
fig, ax = plt.subplots(9,9, figsize = (30,30))

day = []

for idx, comb in enumerate(combination):
    n = list(range(1, 100))
    
    # Make list for Owl and Hawk population with initial value of 150 and 200.
    On = [150] ; Hn = [200]
    
    for i in range(len(n)-1):
        if (1+comb[0])*On[i] - comb[2]*On[i]*Hn[i] < 0 or (1+comb[1])*Hn[i] - comb[3]*On[i]*Hn[i] < 0:
            n = list(range(1,i+2))
            day.append(i+2)
            break
        On.append((1+comb[0])*On[i] - comb[2]*On[i]*Hn[i])
        Hn.append((1+comb[1])*Hn[i] - comb[3]*On[i]*Hn[i])
    
    if col >8:
        col = 0
        row += 1
    ax[row, col].plot(n, On, n, Hn)
    col += 1
```
